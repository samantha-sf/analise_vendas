```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lojas = ['Loja A', 'Loja B', 'Loja C', 'Loja D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}

df = pd.DataFrame(vendas_2022, index=lojas)
df
```

```python
meses = list(vendas_2022.keys())
meses
```

```python
fig, axs = plt.subplots(2,2, figsize=(14,8))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Vendas no período de Janeiro a Dezembro de 2022')

cores = ['darkviolet', 'green', 'darkblue', 'coral']

axs[0,0].plot(df.loc['Loja A', meses])
axs[0,0].set_title('Loja A')

axs[0,1].plot(df.loc['Loja B', meses])
axs[0,1].set_title('Loja B')

axs[1,0].plot(df.loc['Loja C', meses])
axs[1,0].set_title('Loja C')

axs[1,1].plot(df.loc['Loja D', meses])
axs[1,1].set_title('Loja D')

for ax in axs.flat:
  ax.grid()

ymin = 0
ymax = 500

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

plt.show()
```

#Melhorando a vizualização do gráfico

```python
fig, axs = plt.subplots(2,2, figsize=(14,8))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Vendas no período de Janeiro a Dezembro de 2022', fontsize=18)

cores = ['darkviolet', 'green', 'darkblue', 'coral']

for i, ax in enumerate(axs.flat):
  ax.plot(df.loc[df.index[i]], color=cores[i], lw=3)
  ax.set_title(f'Vendas na loja {df.index[i]}', loc='left', fontsize=16)
  ax.set_xlabel('Mês', fontsize=14)
  ax.set_ylabel('Número e vendas', fontsize=14)
  ax.tick_params(labelsize=12)
  ax.grid(color='lightgrey')

ymin = 0
ymax = 500

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

plt.show()
```

```python
def gerar_grafico_paleta(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax = sns.barplot(data=maiores_vendas, y=maiores_vendas.index, x='Total de vendas', orient='h', palette=palette)
  ax.set_title('Total de vendas por loja de Janeiro a Dezembro de 2022', loc='left', fontsize=18)
  ax.set_xlabel('Total de vendas', fontsize=14)
  ax.set_ylabel('')

plt.show()
```

#Gráfico de 4 linhas comparando o desempenho das 4 lojas

```python
meses = list(vendas_2022.keys())
meses
```

```python
sns.set_theme()
sns.set_palette('Dark2')

fig, ax = plt.subplots(figsize=(10,5))
ax = sns.lineplot(df.loc['Loja A', meses], label='Loja A', lw=3)
ax = sns.lineplot(df.loc['Loja B', meses], label='Loja B', lw=3)
ax = sns.lineplot(df.loc['Loja C', meses], label='Loja C', lw=3)
ax = sns.lineplot(df.loc['Loja D', meses], label='Loja D', lw=3)

ax.set_title('Desempenho de vendas das 4 lojas de Janeiro a Dezembro de 2022', loc='left', fontsize=20)
ax.set_xlabel('Mês', fontsize=14)
ax.set_ylabel('Número de vendas', fontsize=14)

ax.legend(title='Lojas', loc='upper right', bbox_to_anchor=(1.18, 1.02))

fig.savefig('Comparação entre as 4 lojas.png', transparent=False, dpi=300, bbox_inches='tight')

plt.show()
```

#Gerando insights
Insights que podem ser gerados com esse gráfico:
- Aumento do número de vendas entre Abril e Maio: as 4 lojas tem um aumento de vendas entre Abril e maio, talvez devido ao período de Páscoa.
- Aumento de vendas em Outubro: apos o mês de Maio todas as lojas apresentam uma queda e iniciam um período de estagnação entre junho e Outubro. Então em Outubro há um grande aumento no número de vendas, provavelmnte devido a comemoração do dia das crianças.
- Desempenho da Loja A: essa loja teve um grande aumento nas vendas entre Outubro e Novembro, porém uma queda entre Novembro e Dezembro, diferente das outras 3 lojas que a partir de Outubro só cresceram no número de vendas. Motivo da queda pode ter sido que, devido a ter vendido tanto entre Outubro e Novembro, não sobrou estoque para Dezembro. Lojas B, C e D ficaram dentro do crescimento esperado nos últimos 3 meses do ano, em que as vendas só aumentam devido ao Dia das Crianças e Natal.
