```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lojas = ['Loja A', 'Loja B', 'Loja C', 'Loja D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}

df = pd.DataFrame(vendas_2022, index=lojas)
df
```

```python
meses = list(vendas_2022.keys())
meses
```

```python
fig, axs = plt.subplots(2,2, figsize=(14,8))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Vendas no per√≠odo de Janeiro a Dezembro de 2022')

cores = ['darkviolet', 'green', 'darkblue', 'coral']

axs[0,0].plot(df.loc['Loja A', meses])
axs[0,0].set_title('Loja A')

axs[0,1].plot(df.loc['Loja B', meses])
axs[0,1].set_title('Loja B')

axs[1,0].plot(df.loc['Loja C', meses])
axs[1,0].set_title('Loja C')

axs[1,1].plot(df.loc['Loja D', meses])
axs[1,1].set_title('Loja D')

for ax in axs.flat:
  ax.grid()

ymin = 0
ymax = 500

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

plt.show()
```

#Melhorando a vizualiza√ß√£o do gr√°fico

```python
fig, axs = plt.subplots(2,2, figsize=(14,8))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Vendas no per√≠odo de Janeiro a Dezembro de 2022', fontsize=18)

cores = ['darkviolet', 'green', 'darkblue', 'coral']

for i, ax in enumerate(axs.flat):
  ax.plot(df.loc[df.index[i]], color=cores[i], lw=3)
  ax.set_title(f'Vendas na loja {df.index[i]}', loc='left', fontsize=16)
  ax.set_xlabel('M√™s', fontsize=14)
  ax.set_ylabel('N√∫mero e vendas', fontsize=14)
  ax.tick_params(labelsize=12)
  ax.grid(color='lightgrey')

ymin = 0
ymax = 500

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

plt.show()
```

#Utilizand a biblioteca Seaborn

```python
import seaborn as sns
```

```python
sns.set_theme()
```

```python
df['Total de vendas'] = df.sum(axis=1)
df
```

```python
maiores_vendas = df.sort_values('Total de vendas', ascending=False)
maiores_vendas
```

```python
sns.barplot(data=maiores_vendas, y=maiores_vendas.index, x='Total de vendas', orient='h')

ax.set(title='Vendas de Janeiro a Dezembro das 4 lojas', xlabel='Total de vendas', ylabel='')

plt.show()
```

```python
def gerar_grafico_paleta(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax = sns.barplot(data=maiores_vendas, y=maiores_vendas.index, x='Total de vendas', orient='h', palette=palette)
  ax.set_title('Total de vendas por loja de Janeiro a Dezembro de 2022', loc='left', fontsize=18)
  ax.set_xlabel('Total de vendas', fontsize=14)
  ax.set_ylabel('')

plt.show()
```

```python
gerar_grafico_paleta('Blues_r')
```

```python
gerar_grafico_paleta('rocket')
```
